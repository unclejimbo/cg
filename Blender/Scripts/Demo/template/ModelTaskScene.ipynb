```python
# this function is used for building emails

def generate_email_content(email, name, time):
    email.mail_content = "Current User: " + getpass.getuser() + "\n" + \
    "Input scene: " + model_dir + '\n' + "Output path: " + path + \
    "/Output/Models/" + name + "/" + scene_name + "/" + "\n" + "GenusNumber: " + str(len(f_list)) + \
    "\n" + "SceneNumber: " + str(scene_number) + "\n" + "Start Time: " + str(start_time) + \
    "\n" + "End Time: " + str(datetime.datetime.now()) + \
    "\n" + "Time Cost: " + str((time - start_time)).split('.')[0] + \
    "\n" + "Time Cost Cell: " + str((time - cell_start_time)).split('.')[0] + "\n" + \
    "Current Filename: " + nb_name + "\n" + "Current IP: " + socket.gethostbyname(socket.gethostname()) + \
    "\n" + "Config: " + "\n" + str(json.dumps(config.__dict__,  sort_keys=True, indent=2)) + \
    "\n" + "Model Name: " + '\n\t\t\t'.join(scene_name_list)

    email.image_path = path + "/Output/Models/" + task_name + "/" + name + "/" + scene_name + "/" + dirname + "/Jpg/image_000.jpg" 
    email.video_path = "/home/zhaoc/cg/Blender/Output/Models/" + task_name + "/" + name + "/" + scene_name + "/" + dirname + "/video/result.mp4"
    return email


def generate_path(cell_name, dirname):
    path_list = []
    output = path + "/Output/Models/" + task_name + "/" + cell_name + "/" + scene_name + "/" + dirname + "/"
    path_list.append(output)
    image_input = "/home/zhaoc/cg/Blender/Output/Models/" + task_name + "/" + cell_name + "/" + scene_name + "/" + dirname + "/"
    path_list.append(image_input)
    video_path = "/home/zhaoc/cg/Blender/Output/Models/" + task_name + "/" + cell_name + "/" + scene_name + "/" + dirname + "/video/result.mp4"
    path_list.append(video_path)
    return path_list
```
